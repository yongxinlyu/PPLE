```python
#pixel art!
#Stardew Valley cat!

clearscreen()
initializeTurtle()
speed(13)

#pixels
def dot(color):
  pendown()
  pencolor(color)
  fillcolor(color)
  begin_fill()
  for _ in range(4):
    forward(20)
    right(90)
  end_fill()
  penup()

#position
size = 20
def goto_grid(x,y):
    penup()
    goto(x*size,y*size)
    pendown()

#locate and draw
def pos(color):
  goto_grid(gx,y)
  dot(color)

#colors
D14 = "#973D9F"
F11 = "#592120"
F12 = "#FD4F6A"
A7 = "#FF8F36"
A6 = "#FDC043"
A4 = "#FFF15D"


y=-2+2+10
for gx in [4,5,11,12]:
  pos(F11)

y=-3+2+10
for gx in [4,6,10,12]:
  pos(F11)
for gx in [5,11]:
  pos(A6)

y=-4+2+10
for gx in [4,7,8,9,12]:
  pos(F11)
for gx in [5,11]:
  pos(F12)
for gx in [6,10]:
  pos(A6)

y=-5+2+10
for gx in [4,12]:
  pos(F11)
for gx in [5,11]:
  pos(F12)
for gx in [6,7,8,9,10]:
  pos(A6)

y=-6+2+10
for gx in [2,3,13,14]:
  pos(F11)
for gx in [4,12]:
  pos(A7)
for gx in [5,6,7,8,9,10,11]:
  pos(A6)

y=-7+2+10
for gx in [3,13]:
  pos(F11)
for gx in [4,5,7,8,9,11,12]:
  pos(A6)
for gx in [6,10]:
  pos(A7)

y=-8+2+10
for gx in [2,3,6,10,13,14]:
  pos(F11)
for gx in [4,5,7,8,9,11,12]:
  pos(A6)

y=-9+2+10
for gx in [3,6,10,13]:
  pos(F11)
for gx in [4,12]:
  pos(A7)
for gx in [5,7,8,9,11]:
  pos(A6)

y=-10+2+10
for gx in [2,3,4,12,13,14]:
  pos(F11)
for gx in [5,11]:
  pos(A7)
for gx in [6,7,9,10]:
  pos(A6)
for gx in [8]:
  pos(F12)

y=-11+2+10
for gx in [5,6,10,11]:
  pos(F11)
for gx in [7,8,9]:
  pos(D14)

y=-12+2+10
for gx in [4,5,7,9,11,12]:
  pos(F11)
for gx in [6,10]:
  pos(A7)
for gx in [8]:
  pos(A4)

y=-13+2+10
for gx in [4,5,8,11,12]:
  pos(F11)
for gx in [6,10]:
  pos(A6)
for gx in [7,9]:
  pos(A7)

y=-14+2+10
for gx in [3,5,11,13]:
  pos(F11)
for gx in [4,6,7,8,9,10,12]:
  pos(A6)

y=-15+2+10
for gx in [3,5,7,9,11,13]:
  pos(F11)
for gx in [4,6,8,10,12]:
  pos(A6)

y=-16+2+10
for gx in [3,5,7,9,11,13]:
  pos(F11)
for gx in [4,8,12]:
  pos(A7)
for gx in [6,10]:
  pos(A6)

y=-17+2+10
for gx in [4,5,7,8,9,11,12]:
  pos(F11)
for gx in [6,10]:
  pos(A6)

y=-18+2+10
for gx in [5,7,9,11]:
  pos(F11)
for gx in [6,10]:
  pos(A6)

y=-19+2+10
for gx in [6,7,9,10]:
  pos(F11)
```
