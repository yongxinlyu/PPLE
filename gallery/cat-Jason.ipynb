```python
clearscreen()
initializeTurtle()
speed(10)
pensize(2)


# 头部 (head)
penup(); goto(0, -30); pendown()
fillcolor("#F2C9A0")  # light beige-orange
begin_fill()
circle(60)
end_fill()

# 左耳朵 (left ear - smoother curve)
penup(); goto(-45, 55); pendown()
fillcolor("#F2C9A0")
begin_fill()
setheading(160)
forward(25)
setheading(70)
forward(40)
setheading(-40)
forward(40)
end_fill()

# 右耳朵 (right ear)
penup(); goto(45, 55); pendown()
fillcolor("#F2C9A0")
begin_fill()
setheading(20)
forward(25)
setheading(110)
forward(40)
setheading(220)
forward(40)
end_fill()

# 耳朵内侧 (inner ears)
fillcolor("#F6A9B0")  # pink
# 左
penup(); goto(-37, 58); pendown()
begin_fill()
setheading(160)
forward(15)
setheading(70)
forward(25)
setheading(-40)
forward(25)
end_fill()
# 右
penup(); goto(37, 58); pendown()
begin_fill()
setheading(20)
forward(15)
setheading(110)
forward(25)
setheading(220)
forward(25)
end_fill()

# 左眼
penup()
goto(-25,50)
pendown()
fillcolor("white")
begin_fill()
circle(10)
end_fill()

# 右眼
penup()
goto(20, 50)
pendown()
begin_fill()
circle(10)
end_fill()

# 左瞳孔
penup()
goto(-25, 50)
pendown()
fillcolor("black")
begin_fill()
circle(4)
end_fill()

# 右瞳孔
penup()
goto(20, 50)
pendown()
begin_fill()
circle(4)
end_fill()


# 鼻子 (nose)
penup(); goto(0, 20); pendown()
fillcolor("#D66A6A")
begin_fill()
setheading(270); forward(6)
setheading(150); forward(6)
setheading(30); forward(6)
end_fill()

# 嘴巴 (mouth)
penup(); goto(0, 20); pendown()
setheading(300); circle(10, 60)
penup(); goto(0, 20); pendown()
setheading(240); circle(-10, 60)

# 胡须 (whiskers)
pencolor("black"); pensize(2)

# 左侧
penup(); goto(-10, 15); pendown()
setheading(180); forward(25)
penup(); goto(-10, 10); pendown()
setheading(200); forward(25)
penup(); goto(-10, 20); pendown()
setheading(160); forward(25)

# 右侧
penup(); goto(10, 15); pendown()
setheading(0); forward(25)
penup(); goto(10, 10); pendown()
setheading(340); forward(25)
penup(); goto(10, 20); pendown()
setheading(20); forward(25)


hideturtle()
```
