```python
initializeTurtle((800,1200))
speed(13)
# ---------- 小工具 ----------
def move(x, y):
    penup(); goto(x, y); pendown()

def set_pen(width=6, color="black"):
    pensize(width); pencolor(color)

def filled_circle(x, y, r, color="black"):
    penup(); goto(x, y - r); setheading(0)
    pendown(); pencolor(color); fillcolor(color)
    begin_fill(); circle(r); end_fill()

def short_hatch(x, y, angle=0, length=14):
    setheading(angle)
    move(x, y)
    forward(length)

# ---------- 完整头部（A） ----------
def draw_chibi_head(x=0, y=0, scale=0.85, line=8, tilt=-30):
    """
    x, y   : 整体中心
    scale  : 缩放（1.0 基准；0.85更小，1.25更大）
    line   : 线条粗细
    tilt   : 整体轻微倾斜（度），正值逆时针
    """
    s = scale
    set_pen(line, "black")

    # 0) 轻微整体倾斜（通过各段的heading加偏移实现）
    T = tilt

    # 2) 左耳
    move(x-165*s, y+22*s); setheading(215+T)
    circle(30*s, 220)

    # 3) 右耳
    move(x+165*s, y-22*s); setheading(-35+T)
    circle(30*s, 220)

    # 顶部到左侧外轮
    setheading(162+T); circle(-340*s, 28)
    setheading(205+T); circle(-340*s, 20)

    # 左侧外轮+小翘
    right(120); forward(42*s); backward(42*s)
    setheading(292+T); circle(210*s, 24)

    # 5) 刘海（3~4撮，略盖住额头）
    # 中央大刘海
    move(x+20*s, y+92*s); setheading(-108+T)
    circle(95*s, 60)
    setheading(66+T); circle(-80*s, 48)

    # 中右刘海
    move(x+70*s, y+84*s); setheading(-112+T)
    circle(72*s, 58)
    setheading(74+T); circle(-60*s, 44)

    # 中左刘海
    move(x-22*s, y+90*s); setheading(-123+T)
    circle(92*s, 64)
    setheading(72+T); circle(-74*s, 42)

    # 额侧碎发（左）
    move(x-85*s, y+84*s); setheading(-140+T)
    circle(78*s, 56)
    setheading(40+T); circle(-64*s, 48)

    # 6) 眼睛（实心圆；间距与高度可按需改）
    filled_circle(x-60*s, y+14*s, 18*s, "black")
    filled_circle(x+60*s, y+14*s, 18*s, "black")

    # 7) 嘴巴（两段小弧，形成轻微“w”）
    set_pen(line-2 if line>4 else line, "black")
    move(x-20*s, y-16*s); setheading(-12+T)
    circle(44*s, 30)
    setheading(205+T); circle(-44*s, 30)

    # 8) 脸颊短划线（腮红感）
    set_pen(line-3 if line>5 else max(2, line-1), "black")
    short_hatch(x-102*s, y+4*s, angle=22+T,  length=20*s)
    short_hatch(x-110*s, y+14*s, angle=27+T, length=16*s)
    short_hatch(x+102*s, y+4*s, angle=158+T, length=20*s)
    short_hatch(x+110*s, y+14*s, angle=153+T, length=16*s)

# ---------- 一键绘制 ----------
clear()
draw_chibi_head(x=0, y=0, scale=1.12, line=8, tilt=-30)
```
