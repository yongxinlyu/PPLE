```python
# Draw a cat!

clearscreen()
initializeTurtle()
speed(9)

forward(40) #head
left(60)
forward(30)
right(120)
forward(80)
right(60)
forward(50)
right(60)
forward(100)
right(60)
forward(50)
right(60)
forward(80)
right(120)
forward(30)

penup() #left eye
goto(-5,-25)
pendown()
setheading(270)
forward(15)

penup() #right eye
goto(45,-25)
pendown()
setheading(270)
forward(15)

penup() #mouth
goto(20,-60)
pendown()
setheading(-60)
circle(10,120)
penup()
goto(20,-60)
pendown()
setheading(-120)
circle(-10,120)

penup() #left wisker 1
goto(-35,-50)
pendown()
setheading(170)
forward(30)

penup() #left wisker 2
goto(-35,-55)
pendown()
setheading(180)
forward(30)

penup() #left wisker 3
goto(-35,-60)
pendown()
setheading(190)
forward(30)

penup() #right wisker 1
goto(75,-50)
pendown()
setheading(10)
forward(30)

penup() #right wisker 2
goto(75,-55)
pendown()
setheading(0)
forward(30)

penup() #right wisker 3
goto(75,-60)
pendown()
setheading(350)
forward(30)

def heart(size,color): #reusable heart
  pencolor(color)
  fillcolor(color)
  begin_fill()
  left(150)
  forward(9*size)
  circle(-9*size,200)
  left(120)
  circle(-9*size,200)
  forward(18*size)
  end_fill()

penup() #heart
goto(20,20)
pendown()
pencolor("red")
fillcolor("red")
heart(1,"red")

def star(size, color): #reusable star
  pencolor(color)
  fillcolor(color)
  begin_fill()
  for _ in range(5):
    forward(size)
    right(144)
    forward(size)
    left(72)
  end_fill()


for i in range(12): #circle of stars
  penup()
  goto(20,-30)
  setheading(i*30+90)
  forward(120)
  pendown()
  left(54)
  star(15,"yellow")



hideturtle()
```
