### Word Frequency

This notebook extracts high-frequency words from a YouTube presentation transcript.
It is part of my personal project to improve presentation skills by identifying and learning productive vocabulary and phrases.

```python
import re

infile = open("../data/rawscript.txt","r")
lines = infile.readlines()
infile.close()

words = []
single_list = {}
double_list = {}
triple_list = {}
quad_list = {}
penta_list = {}
hexa_list = {}
```

#### Trim and format script 

```python
def main():
    formatscript()
    singlelist()
    doublelist()
    triplelist()
    quadlist()
    pentalist()
    hexalist()
    
def formatscript():
    for line in lines:
        line_word = line.lower().replace('\n','').split(" ");
        for w in line_word:
            if re.match("\d*[:]\d",w):
                continue
            else:
                words.append(w)

    outfile=open("../data/processedscript.txt","w")

    for i in range(0,len(words)):
        outfile.write(words[i]+' ')
    outfile.close()
```

#### Print single word list according to occurrence

```python
def singlelist():
    for word in words:
        if word in single_list:
            single_list[word] += 1
        else:
            single_list[word] = 1

    print(sorted(single_list.items(), key=lambda kv:kv[1],reverse=True))
```

#### Print double word list according to occurrence

```python
def doublelist():
    for i in range(0, len(words)-1):
        two_words = words[i] + " " + words[i+1]
        if two_words in double_list:
            double_list[two_words] += 1
        else:
            double_list[two_words] = 1


    print(sorted(double_list.items(), key=lambda kv:kv[1],reverse=True))
```

#### Print triple word list according to occurrence

```python
def triplelist():
    for i in range(0, len(words)-2):
        three_words = words[i] + " " + words[i+1] + " " + words[i+2]
        if three_words in triple_list:
            triple_list[three_words] += 1
        else:
            triple_list[three_words] = 1


    print(sorted(triple_list.items(), key=lambda kv:kv[1],reverse=True))

```

#### Print quad word list according to occurrence

```python
def quadlist():
    for i in range(0, len(words)-3):
        four_words = words[i] + " " + words[i+1] + " " + words[i+2]+ " " + words[i+3]
        if four_words in quad_list:
            quad_list[four_words] += 1
        else:
            quad_list[four_words] = 1


    print(sorted(quad_list.items(), key=lambda kv:kv[1],reverse=True))
```

#### Print penta word list according to occurrence

```python
def pentalist():
    for i in range(0, len(words)-4):
        five_words = words[i] + " " + words[i+1] + " " + words[i+2] + " " + words[i+3] + " " + words[i+4]
        if five_words in penta_list:
            penta_list[five_words] += 1
        else:
            penta_list[five_words] = 1


    print(sorted(penta_list.items(), key=lambda kv:kv[1],reverse=True))
```

#### Print hexa word list according to occurrence

```python
def hexalist():
    for i in range(0, len(words)-5):
        six_words = words[i] + " " + words[i+1] + " " + words[i+2] + " " + words[i+3] + " " + words[i+4] + " " + words[i+5]
        if six_words in hexa_list:
            hexa_list[six_words] += 1
        else:
            hexa_list[six_words] = 1


    print(sorted(hexa_list.items(), key=lambda kv:kv[1],reverse=True))
```

```python
if __name__ == '__main__':
    main()
```
